```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
from scipy import stats
%matplotlib inline
plt.style.use('seaborn-dark')
```

```python
beer_totals = pd.read_csv(
    '../data/../data/minor_beers.csv', 
    delimiter=","
)
beer_totals['name'] = beer_totals['Line Item']
beer_totals.set_index('name', inplace=True)
beer_totals.drop("Unnamed: 0", axis = 1, inplace=True)
print(beer_totals.shape)
beer_totals.head(1)
```

```python
beer_sales = pd.read_csv(
    '../data/../data/minor_beer_sales.csv', 
    delimiter=","
)
beer_sales.set_index("Time", inplace=True)
beer_sales["Time"] = beer_sales["Time2"]

beer_sales["datetime"] = pd.to_datetime(beer_sales["Time"])
beer_sales["datetime_hour"] = pd.DatetimeIndex(beer_sales['Time']).round("h")
beer_sales['year'] = pd.DatetimeIndex(beer_sales['datetime']).to_period("Y")
cols_to_drop = ["Time.1", "Time2", ]

beer_sales['name'] = beer_sales['Line Item'] + " " + beer_sales['year'].astype('str')
beer_sales.drop(cols_to_drop, inplace=True, axis=1)
print(beer_sales.shape)
beer_sales.head(1)
```

```python
beer_sales["Price"].unique()
```

## Let's start by graphing the minor beers, each on its own chart

```python
fig, ax = plt.subplots(len(beer_sales["name"].unique()),1, figsize=(20,4*len(beer_totals.index)))
i= 0
for beer in beer_sales["name"].unique():
    this_beer = beer_sales[beer_sales["name"] == beer]
    sales_by_hour = this_beer.groupby(this_beer["datetime_hour"].dt.date).sum()
    
    this_beer.head(2)
    ax[i].plot("Quantity", data=sales_by_hour, linewidth=2, alpha=1, label=beer)

    ax[i].set_title(f"Number of { beer } sold.")
    ax[i].set_ylabel("Number of Sales")
    ax[i].legend()

    i += 1
```

## What happens if they all start on the same day?

```python
fig, ax = plt.subplots(1, figsize=(20,5))
max_list_size = 0
beer_series = []

def zerolistmaker(n):
    listofzeros = [0] * n
    return listofzeros
x =zerolistmaker(5)
y =zerolistmaker(5)
x.extend(y)
print(x)


for beer in beer_sales["name"].unique():

    this_beer = beer_sales[beer_sales["name"] == beer].copy()
    
    this_beer['hour_ordinal'] = this_beer['datetime_hour'].apply(lambda x: x.toordinal())
    minimum_ord = this_beer['hour_ordinal'].min()
    this_beer['hour_ordinal_normalized'] = this_beer['hour_ordinal']  - minimum_ord

    if this_beer['hour_ordinal_normalized'].max() < 150:
        sales_by_hour = this_beer.groupby(this_beer["datetime_hour"].dt.date).sum()
        list_of_sales = this_beer['Quantity'].values.astype(int).tolist()
        if len(list_of_sales) < 175:
            z = zerolistmaker(175-len(list_of_sales))
            list_of_sales.extend(z)
#             print(len(list_of_sales))
        
        if len(list_of_sales) > max_list_size:
            max_list_size = len(list_of_sales)
        beer_series.append(list_of_sales)
        
        
# print(beer_series)
ax.stackplot(np.arange(0,max_list_size), *beer_series, linewidth=2, alpha=1, baseline='zero')

ax.set_title(f"Trends in Beer sales")
ax.set_ylabel("Number of Sales")
```

## Build a linear regression line for each beer, and normalize the times, and y intercept so that they all start at the same place

```python
fig, ax = plt.subplots(1, figsize=(20,4))
# Save the slopes for a later cell
slopes=[]

for beer in beer_sales["name"].unique():
    
    this_beer = beer_sales[beer_sales["name"] == beer].copy()
    
    this_beer['hour_ordinal'] = this_beer['datetime_hour'].apply(lambda x: x.toordinal())
    minimum_ord = this_beer['hour_ordinal'].min()
    this_beer['hour_ordinal_normalized'] = this_beer['hour_ordinal']  - minimum_ord
    sales_by_hour = this_beer.groupby(this_beer["datetime_hour"].dt.date).sum()

    if this_beer['hour_ordinal_normalized'].max() < 150:
        m, b = np.polyfit(this_beer['hour_ordinal_normalized'],this_beer['Quantity'],1)
        slopes.append(m)
        # Normally we would leave a "+b" in the next part, but in this case we are only interested in the slopes.
        ax.plot(this_beer['hour_ordinal_normalized'], m*this_beer['hour_ordinal_normalized'])
    
    sales_by_hour = this_beer.groupby(this_beer["datetime_hour"].dt.date).sum()

ax.set_title(f"Regression slopes for each specialty beer")
ax.set_ylabel("Slope of Regression")
```

### If we take a minute and look at the values sorted, we see that it seems to form a symetrical distribution.

```python
slopes.sort()
x_vals = np.arange(0, len(slopes))

fig, ax = plt.subplots(1,1, figsize=(20,4))
ax.set_title(f"Distribution of regression slopes for specialty beers")
ax.set_ylabel("Slope of Regression")

ax.plot(x_vals, slopes, linewidth=2, alpha=1, label=beer)
```

## Let's look at the distance from the median for each slope

```python
slopes = np.array(slopes)
slope_distances = []
slopes_mean = slopes.mean()
print(f"Slopes mean is {slopes_mean}")
for slope in slopes:
    distance_to_mean = abs(slope-slopes_mean)
    slope_distances.append(distance_to_mean)

x_vals = np.arange(0, len(slopes))

fig, ax = plt.subplots(1,1, figsize=(20,4))
ax.set_title(f"Distribution of regression slopes for specialty beers")
ax.set_ylabel("Slope of Regression")

ax.plot(x_vals, slope_distances, linewidth=2, alpha=1, label=beer)
```

### Creating a histogram indicates something close to a normal distribution

```python
slopes = np.array(slopes)
slopes_std = slopes.std()
slopes_mean = slopes.mean()

fig, ax = plt.subplots(4,1, figsize=(20,6))

slopes_dist = stats.norm(loc=slopes_mean, scale=slopes_std)
support = np.linspace(-0.04, .03, 100)

x_vals = np.arange(0, len(slopes))
ax[0].hist(slopes, linewidth=2, alpha=1, label=beer, bins=9)
ax[1].plot(support, slopes_dist.pdf(support))

ax[2].hist(slopes, linewidth=2, alpha=1, label=beer, bins=9)
ax[2].plot(support, slopes_dist.pdf(support))

ax[3].hist(slopes, linewidth=2, alpha=1, label=beer, bins=50)
ax[3].plot(support, slopes_dist.pdf(support)/9)


print(f" There are { len(slopes) } beers being studied here.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)

# Use only one feature
diabetes_X = diabetes_X[:, np.newaxis, 2]

# Split the data into training/testing sets
diabetes_X_train = diabetes_X[:-20]
diabetes_X_test = diabetes_X[-20:]

# Split the targets into training/testing sets
diabetes_y_train = diabetes_y[:-20]
diabetes_y_test = diabetes_y[-20:]

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_X_train, diabetes_y_train)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(diabetes_X_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(diabetes_y_test, diabetes_y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(diabetes_y_test, diabetes_y_pred))

# Plot outputs
plt.scatter(diabetes_X_test, diabetes_y_test,  color='black')
plt.plot(diabetes_X_test, diabetes_y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
